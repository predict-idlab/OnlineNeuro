```python
import sys
from pathlib import Path
sys.path.append(str(Path('../..') / 'cajal'))
sys.path.append(str(Path('..') / 'simulators'))

import matplotlib.pyplot as plt
import numpy as np

from cajal.nrn.stimuli import Stimulus
from cajal.units.utils import unitdispatch, strip_units
import unyt
```

```python
#simulators/python/pulses/pulse_definitions.py
from python.pulses import PulseRamp
```

```python
pulse0 = PulseRamp(
    delay=1,
    amplitude=2,
    pulse_width=4,
    interphase_gap=0.5,
    decay_width=2,
    k=0.90,
    tau=10,
    ramp_width=2,
    offset=0,
)
pulse1 = PulseRamp(
    delay=1,
    amplitude=2,
    pulse_width=4,
    interphase_gap=0.5,
    decay_width=2,
    k=1,
    ramp_width=2,
    offset=0,
)
```

```python
# Example usage
t = np.linspace(0, 10, 1000)
t = unyt.unyt_array(t, 'ms')
signal0 = pulse0.timecourse(t)
signal1 = pulse1.timecourse(t)
signal2 = 2*pulse0.timecourse(t)

plt.plot(t, signal0)
plt.plot(t, signal1)
plt.plot(t, signal2)

plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Pulse Exp Ramp Signal")
plt.grid(True)
plt.show()
```
