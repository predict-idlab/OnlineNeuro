```python
import matplotlib.pyplot as plt
import numpy as np
from simulators.python.pulses import PulseRamp
import unyt
```

```python
pulse0 = PulseRamp(
    delay=0,
    amplitude=2,
    pulse_width=3,
    interphase_gap=1,
    decay_width=2,
    k=0.90,
    tau=10,
    ramp_width=3,
    offset=0,
)
pulse1 = PulseRamp(
    delay=1,
    amplitude=2,
    pulse_width=4,
    interphase_gap=0.5,
    decay_width=2,
    k=1,
    ramp_width=2,
    offset=0,
)
```

```python
# Example usage
t = np.linspace(0, 10, 1000)
t = unyt.unyt_array(t, 'ms')
signal0 = pulse0.timecourse(t)
signal1 = pulse1.timecourse(t)
signal2 = 2*pulse0.timecourse(t)

plt.plot(t, signal0, label='Pulse 1')
plt.plot(t, signal1, label='Pulse 2')
plt.plot(t, signal2, label='Pulse 3')

plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("Pulse Exp Ramp Signal")
plt.legend()
plt.grid(True)
plt.show()
```
